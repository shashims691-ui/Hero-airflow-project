**SILVER LAYER (CLEANED DATA)**

```python
# Create copies for Silver Layer

orders_silver = orders.copy()
customers_silver = customers.copy()
order_items_silver = order_items.copy()
payments_silver = payments.copy()
reviews_silver = reviews.copy()
products_silver = products.copy()
sellers_silver = sellers.copy()

print("Silver Layer Created Successfully")
```

```python
orders_silver['order_purchase_timestamp'] = pd.to_datetime(
    orders_silver['order_purchase_timestamp']
)

orders_silver['order_approved_at'] = pd.to_datetime(
    orders_silver['order_approved_at']
)

orders_silver['order_delivered_carrier_date'] = pd.to_datetime(
    orders_silver['order_delivered_carrier_date']
)

orders_silver['order_delivered_customer_date'] = pd.to_datetime(
    orders_silver['order_delivered_customer_date']
)

orders_silver['order_estimated_delivery_date'] = pd.to_datetime(
    orders_silver['order_estimated_delivery_date']
)

print("Date Conversion Completed")
```

```python
orders_silver.info()
```

```python
orders_silver['delivery_days'] = (
    orders_silver['order_delivered_customer_date']
    - orders_silver['order_purchase_timestamp']
).dt.days

orders_silver['delay_days'] = (
    orders_silver['order_delivered_customer_date']
    - orders_silver['order_estimated_delivery_date']
).dt.days

print("Delivery KPIs Created")
```

```python
orders_silver[['delivery_days','delay_days']].head()
```

**Seller Performance Table**

```python
seller_performance = order_items_silver.merge(
    fact_orders[['order_id','delay_days']],
    on='order_id',
    how='left'
)

seller_performance = seller_performance.groupby('seller_id').agg({
    'price':'sum',
    'freight_value':'sum',
    'delay_days':'mean'
}).reset_index()

seller_performance.rename(columns={
    'price':'total_sales',
    'freight_value':'total_freight',
    'delay_days':'avg_delay_days'
}, inplace=True)

seller_performance.head()
```

**Top 10 Sellers**

```python
top_sellers = seller_performance.sort_values(
    by='total_sales',
    ascending=False
)

top_sellers.head(10)
```
